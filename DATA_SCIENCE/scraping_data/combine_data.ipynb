```python
import pandas as pd
```

```python
df1 = pd.read_csv('../full_dataset_glints_v2.csv')
df1.info()
```

```python
# Secara default, keep='first' artinya kemunculan pertama dianggap unik, kemunculan berikutnya dianggap duplikat
total_duplikat = df1.duplicated(subset=["Job_Link"]).sum()

print(f"Total baris duplikat berdasarkan Job_Link: {total_duplikat}")
```

```python
df2 = pd.read_csv('postingan_terbaru.csv')
```

```python
df2.info()
```

```python
# Secara default, keep='first' artinya kemunculan pertama dianggap unik, kemunculan berikutnya dianggap duplikat
total_duplikat = df2.duplicated(subset=["Job_Link"]).sum()

print(f"Total baris duplikat berdasarkan Job_Link: {total_duplikat}")

df2.drop_duplicates(subset=["Job_Link"], inplace=True)

# total baris setelah menghapus duplikat
total_baris_setelah_hapus_duplikat = len(df2)
print(f"Total baris setelah menghapus duplikat: {total_baris_setelah_hapus_duplikat}")

# total duplikat setelah menghapus duplikat
total_duplikat_setelah_hapus = df2.duplicated(subset=["Job_Link"]).sum()
print(f"Total duplikat setelah menghapus duplikat: {total_duplikat_setelah_hapus}") 
```

```python
# 1. Filter df2 untuk mengambil data yang Job_Link-nya BELUM ADA di df1
data_baru = df2[~df2["Job_Link"].isin(df1["Job_Link"])]

# 2. Gabungkan df1 dengan data_baru tersebut
df_gabungan = pd.concat([df1, data_baru], ignore_index=True)
```

```python
# melihat banyak data baru

print(f"Banyak data baru saja didapat adalah: {len(data_baru)}")
```

```python
df_gabungan.info()
```

```python
# melihat jumlah job_link yang unik setelah digabungkan
jumlah_job_link_unik = df_gabungan['Job_Link'].nunique()
print(f"Jumlah Job_Link unik setelah digabungkan: {jumlah_job_link_unik}")
```

```python
# Secara default, keep='first' artinya kemunculan pertama dianggap unik, kemunculan berikutnya dianggap duplikat
total_duplikat = df_gabungan.duplicated(subset=["Job_Link"]).sum()

print(f"Total baris duplikat berdasarkan Job_Link: {total_duplikat}")
```
